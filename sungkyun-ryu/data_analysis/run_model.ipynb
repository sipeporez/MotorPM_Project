```python
def PostProcessing(file_name, cols=['spectrum_x_amp', 'spectrum_y_amp', 'spectrum_z_amp'], name="spectrum_total"):
    """ csv 파일에 대한 전처리 함수
 
    file_name : 파일 경로
    cols : 컬럼 리스트 ex) ['spectrum_x_amp', 'spectrum_y_amp', 'spectrum_z_amp']
    name : 컬럼 이름
    """
    # csv 파일 로드
    df = pd.read_csv(file_name, header=[0], index_col=[0, 1])
    # 각 행의 리스트를 병합
    for col in tqdm.tqdm(cols):
        df[col] = df[col].apply(lambda x: [float(j) for j in x.split(',')])
    new_df = df[cols].apply(lambda x: list(chain.from_iterable(x)), axis=1)
 
    # 데이터프레임으로 변환
    result = new_df.groupby(level=0).apply(lambda x: [item for sublist in x for item in sublist]).to_frame(name=name)
    print(type(result),type(result[name]),len(result[name]))
    return result
```
