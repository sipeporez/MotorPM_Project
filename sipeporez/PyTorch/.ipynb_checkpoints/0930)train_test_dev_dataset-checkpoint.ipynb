```python
df = pd.read_csv('data/5528_droped_data.csv')
```

```python
target = df.loc[:, ['imbalance_health', 'misalignment_health', 'looseness_health', 'bearing_health']]
```

```python
# 각 열의 문자열을 쉼표로 구분하여 float list로 변환
df['spectrum_x_amp'] = df['spectrum_x_amp'].apply(lambda x: ','.join(map(str, x)))
df['spectrum_y_amp'] = df['spectrum_y_amp'].apply(lambda x: ','.join(map(str, x)))
df['spectrum_z_amp'] = df['spectrum_z_amp'].apply(lambda x: ','.join(map(str, x)))
```

```python
grouped_df = df.groupby(df.index // 12).apply(lambda x: x.reset_index(drop=True))
```
